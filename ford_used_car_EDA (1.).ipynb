### Create a function to find the number of observations for unique value in a column of a DataFrame

```python
def unique_value_count(data, col):
    col_count = data[col].value_counts(normalize= True)
    col_count = pd.DataFrame(col_count)        
    col_count = col_count.reset_index()          
    col_count = col_count.rename(columns = {'index': col, col: '% of Cars'})   
    col_count['% of Cars'] = np.round(col_count['% of Cars']*100, 2)
    return col_count
```

```python
model_count = unique_value_count(df, 'model')
display(model_count)
```

```python
year_count = unique_value_count(df, 'year')
#display(year_count)
# sort the year_count DataFrame based on 'year'
#year_count = year_count.sort_values(by = 'year')
display(year_count)
```

```python
transmission_count = unique_value_count(df, 'transmission')
display(transmission_count)
```

```python
fuelType_count = unique_value_count(df, 'fuelType')
display(fuelType_count)
```

### Create a function to show the number of observations for each unique value in a column using a barplot

```python
def barplot(data, x_col, y_col, color, rotation,yticks):
    sns.barplot(x = x_col, y = y_col, data = data, color = color, alpha = 0.75)
    plt.xlabel( x_col)
    plt.ylabel('No. of Cars')
    plt.title('No. of Cars present for each'  +   x_col)
    plt.yticks(yticks)
    plt.xticks(rotation = rotation)
```

```python
sns.set_context("talk")

plt.figure(figsize = (20,12))
plt.subplot(2,2,1)
barplot(model_count,'model', '% of Cars', 'green', 90, np.arange(0,41,10))

plt.subplot(2,2,2)
barplot(year_count,'year', '% of Cars', 'red', 90, np.arange(0,31,10) )

plt.subplot(2,2,3)
barplot(transmission_count,'transmission', '% of Cars', 'orange', 0, np.arange(0,91,10))

plt.subplot(2,2,4)
barplot(fuelType_count,'fuelType', '% of Cars', 'blue', 0, np.arange(0,71,10))

plt.suptitle('Categorical columns in the Ford used Cars')
plt.tight_layout()
plt.show()
```

### Using Scatter plot, find the relation between numerical columns of the dataset:
* mileage vs. price
* mpg vs. price

```python
sns.set_context('talk')
plt.figure(figsize=(20,12))
plt.subplot(2,1,1)
sns.scatterplot(x='mileage', y='price', data=df)
plt.title("row = 1, column = 1, position = 1")


plt.subplot(2,1,2)
sns.scatterplot(x='mpg', y='price', data=df)
plt.title("row = 2, column = 1, position = 2")

plt.suptitle('Numerical columns in the Ford used cars by scatter plot')

plt.tight_layout()
plt.show()
```

#### From these scatter plots, we can conclude that:
 * **mileage vs price**: when the mileage of the vehicle increases, the price of the vehicle decreases.
 * **mpg vs price**: when the mpg of the vehicle increases, the price of the vehicle also increases. But this is not the case always. So we can not consider this relation to predict the price of the vehicle.

### Changing the 'hue' parameter and  creating a 'Trend line'

```python
plt.figure(figsize=(20,12))
plt.subplot(2,1,1)
sns.scatterplot(x='mileage', y='price', data=df, hue= 'transmission')
plt.title("row = 1, column = 1, position = 1")

plt.subplot(2,1,2)
sns.regplot(x='mileage', y='price', data=df, scatter_kws={'color':'red','edgecolor': 'white'})
plt.title("row = 2, column = 1, position = 2")

plt.tight_layout()
plt.show()
```

#### Interpretation:
 * Adding 'transmission' as a 'hue' parameter of the first graph, we can easily differentiate that which car has more mileage and also identify the price of the vehicle.
 * But using 'Trendline' in the second graph, we can mislead the infromation. Because, the price of the vehicle will never be negative.

### Create Histogram with numerical columns of the dataset

```python
plt.figure(figsize=(20,12))
plt.subplot(2,1,1)
sns.histplot(df['mileage'],binwidth=5000,color= 'green', fill= False)
plt.title("row = 1, column = 1, position = 1")
plt.subplot(2,1,2)
sns.histplot(df['mpg'],binwidth=10,color= 'red')
plt.title("row = 2, column = 1, position = 2")

plt.tight_layout()
plt.show()
```

### Create Distribution plot with  numerical columns of the dataset

```python
plt.figure(figsize=(20,12))
plt.subplot(2,1,1)
sns.histplot(df['mileage'], kde= True,color= 'green')
plt.title("row = 1, column = 1, position = 1")
plt.subplot(2,1,2)
sns.histplot(df['mpg'],kde= True,color= 'red')
plt.title("row = 2, column = 1, position = 2")

plt.tight_layout()
plt.show()
```

### Create ECDF plot with numerical columns of the dataset

```python
plt.figure(figsize=(20,8))
plt.subplot(3,1,1)

sns.histplot(df['mileage'],kde=True,color= 'blue', cumulative= True)
plt.title("row = 1, column = 1, position = 1")

plt.subplot(3,1,2)
sns.histplot(df['mileage'],color= 'blue', kde=True,cumulative= True, stat= 'probability')
plt.title("row = 2, column = 1, position = 2")

plt.tight_layout()
plt.show()
```
